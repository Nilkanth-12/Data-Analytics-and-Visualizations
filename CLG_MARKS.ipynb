```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("C:/Users/nilka/OneDrive/Desktop/CLG_MARKS.csv")
df1.head()
```

```python
df1.isnull()
```

```python
 df2 = df1.replace('AB', np.nan)
 df2
    
```

```python
df3 = df2.dropna(how='all', subset=['DS', 'DBMS', 'FOC-1', 'DL'])
df3
```

```python
df3[['DS', 'DBMS', 'FOC-1', 'DL']] = df3[['DS', 'DBMS', 'FOC-1', 'DL']].apply(pd.to_numeric, errors='coerce')
df3
```

```python
df3['Total'] = df3[['DS', 'DBMS', 'FOC-1', 'DL']].sum(axis=1)
df3
```

```python
df4 = df3[['enr no', 'Total']]
df4
```

```python
plt.scatter(x = df4['enr no'], y = df4['Total'])
plt.xlabel("enr no")
plt.ylabel("Total")
plt.title("Scatter Plot of Student Marks")
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.bar(df4['enr no'], df4['Total'])
plt.xlabel("enr no")
plt.ylabel("Total")
plt.title("Bar Graph of Student Marks")
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(df4['enr no'], df4['Total'], marker='o', color='b')
plt.xlabel("enr no")
plt.ylabel("Total")
plt.title("Line Chart of Student Marks")
plt.grid(True)
plt.show()
```

```python
from scipy import stats

plt.figure(figsize=(15, 6))
plt.scatter(df4['enr no'], df4['Total'], color='blue', label='Data Points')
slope, intercept, r_value, p_value, std_err = stats.linregress(df4['enr no'], df4['Total'])
plt.plot(df4['enr no'], slope * df4['enr no'] + intercept, color='red', label='Trend Line')
plt.xlabel("enr no")
plt.ylabel("Total")
plt.title("Scatter Plot with Trend Line of Student Marks")
plt.legend()
plt.show()
```

```python
threshold = 35
above_threshold = df4[df4['Total'] > threshold].shape[0]
total_students = df4.shape[0]
percentage_above_threshold = (above_threshold / total_students) * 100

plt.figure(figsize=(4, 2))
plt.bar(['Above Threshold', 'Below Threshold'], [percentage_above_threshold, 100 - percentage_above_threshold], color=['green', 'red'])
plt.title(f"Percentage of Students Above and Below {threshold} Marks")
plt.ylabel("Percentage")
plt.show()
```
